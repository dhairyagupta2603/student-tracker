```python
import os
import pathlib
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
os.getcwd()
```

```python
imgs_dir = os.path.join('..', 'data', 'images')
imgs_dir
```

```python
img = cv2.imread(os.path.join(imgs_dir, 'test.jpg'))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
height, width = 480, 640
img_scaled = cv2.resize(img, (width, height), interpolation = cv2.INTER_AREA)
img_scaled.shape
```

```python
plt.imshow(img_scaled)
plt.show()
```

```python
img_scaled = cv2.cvtColor(img_scaled, cv2.COLOR_RGB2BGR)
```

```python
cv2.imwrite(os.path.join(imgs_dir, 'test_pre.jpg'), img_scaled)
```
